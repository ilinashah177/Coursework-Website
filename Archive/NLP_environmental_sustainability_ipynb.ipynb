```python
raw = parser.from_file("China-Pakistan_CGEP_Report_111722.pdf")
```

```python
raw['metadata']
```

```python
date=raw['metadata']['dcterms:created']
title=raw['metadata']['dc:title']
raw_text=raw['content']

print('Date: ', date)
print('Title: ', title)
print('Word Count: ', len(raw_text))
print('Text:')
raw_text
```

```python
word_count = len(raw_text.split())
print("word count:", word_count)
```

```python
import re

def clean_text(raw_text):
    text = re.sub(r'\s+', ' ', raw_text)

    text = re.sub(r'http[s]?://\S+', '', text)
    text = re.sub(r'www\.\S+', '', text)

    text = re.sub(r'\s+', ' ', text)

    text = text.strip()

    return text

cleaned_text = clean_text(raw_text)

print("Word count before cleaning:", len(raw_text))
print("Word count after cleaning:", len(cleaned_text))
print("Preview after cleaning:", cleaned_text[:1000], "...")

```

```python
doc = nlp(cleaned_text)
```

```python
# Extract the first sentence from cleaned_text
sent = list(doc.sents)[0]
print(sent)
```

It is evident that this is the cover page of the paper, containing the title and some basic information, such as the author, Erica Downs, and the publication date, October 2019. The paper is part of the Columbia University Center on Global Energy Policy’s project called "Research on China’s Energy and Climate," and it is one of the series of papers discussing the Belt and Road Initiative.

```python
import nltk
nltk.download('stopwords')
```

```python

from nltk.corpus import stopwords
from collections import Counter

stop_words = set(stopwords.words('english'))

words = []
for token in doc:
    if token.is_alpha and token.lower_ not in stop_words:
        words.append(token.lower_)

print("Total number of valid words:", len(words))
```

```python
freq = Counter(words)
print("The 20 most common words：")
for w, c in freq.most_common(20):
    print(w, c)
```

Among these most frequently used words, the term "coal" caught my attention. So, let's take a closer look at the paragraphs that mention "coal" to understand what they specifically discuss.

```python
# Find and print the sentences in cleaned_text that contain the word "coal."
for sent in doc.sents:
    if "coal" in sent.text.lower():
        print(sent)
        print("----------------")
```

Although there are many sentences containing "coal," from the first few sentences, we learn that nearly 75 percent of the generation capacity of CPEC power plants is coal-fired. This is likely the source of international concerns about the potential environmental issues caused by China’s investment in Pakistan's power projects. Additionally, from these sentences, we learn that China regards some of the CPEC coal power plants as environmentally friendly because they use relatively modern technologies and are expected to emit fewer greenhouse gases compared to the fuel oil plants Pakistan is replacing.

Next, we will delve into the topic of interest for our group: electricity.

```python
# Print the sentences that contain the word "electricity"
for sent in doc.sents:
    if "electricity" in sent.text.lower():
        print(sent)
        print("----------------")
```

One of these sentences caught our attention: "Ultimately, it is up to the host country to decide the composition of its electricity mix."
To gather more information related to this, we will also print the 10 sentences before and after this sentence.

```python
# Find the sentence "Ultimately, it is up to the host country to decide the composition of its electricity mix."
# and print the 10 sentences before and after it as context.
for sent in doc.sents:
    if "it is up to the host country to decide the composition of its electricity mix" in sent.text.lower():
        # Print the 10 sentences before this sentence
        for prev_sent in list(doc.sents)[list(doc.sents).index(sent)-10:list(doc.sents).index(sent)]:
            print(prev_sent)
        print("----------------")

        # Print the 10 sentences after this sentence
        for next_sent in list(doc.sents)[list(doc.sents).index(sent)+1:list(doc.sents).index(sent)+11]:
            print(next_sent)
        print("----------------")
```

After reading this contextual information, we understand the current state of Pakistan's composition of its electricity mix. Although influenced by China's initiatives, it is ultimately determined by Pakistan's needs. This indicates that while coal-fired power is viewed as an unclean energy source globally, it is currently what Pakistan requires. Furthermore, in Pakistan, coal-fired power plants are expected to emit fewer greenhouse gases compared to the fuel oil plants they are replacing.

```python
from collections import defaultdict

# Count the number of each entity type
entity_counts = defaultdict(int)
for ent in doc.ents:
    entity_counts[ent.label_] += 1

print("Named entity type statistics:")
for label, count in entity_counts.items():
    print(label, count)
```

```python
from collections import Counter

# Count the occurrences of GPE (Geopolitical Entity) entities
gpe_counts = Counter()
for ent in doc.ents:
    if ent.label_ == "GPE":
        gpe_counts[ent.text] += 1

# Print the top 10 most frequent GPE entities
print("\nTop 10 GPE entities and their frequencies:")
for location, count in gpe_counts.most_common(10):
    print(f"{location}: {count} times")
```

```python
print("Whole Text Polarity:", doc._.blob.polarity)
print("Whole Text Subjectivity:", doc._.blob.subjectivity)
```

Through the analysis of polarity and subjectivity, it can be observed that most of the text in this paper is based on factual information, with low subjectivity. The polarity is close to neutral, and the emotional expression is not intense. This aligns with our general understanding of academic papers.
